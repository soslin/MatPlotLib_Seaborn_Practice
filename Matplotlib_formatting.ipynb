```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
housing_raw = pd.read_csv(
    '../Data/housing_data.csv',
    parse_dates = ['period_begin', 'period_end'],
    dtype = {'total_homes_sold' : 'Int32', 'region_name' : 'category'} 
) #category cuts down on processing time
```

```python
housing_raw.head()
```

```python
housing_raw.info()
```

```python
ca_housing = housing_raw.loc[
    (housing_raw['region_name'].str.contains('CA'))
].assign(region_name = housing_raw['region_name'].str[0:-11])
```

```python
ca_housing.info(memory_usage = 'deep')
```

```python
ca_housing.head()
```

```python
ca_housing_pivot = ca_housing.pivot_table(index='period_begin',
                                         columns = 'region_name',
                                         values = 'median_active_list_price').assign(
    CA_average = lambda x: x.mean(axis = 1))

ca_housing_pivot.head()
```

```python
#Don't do!
ca_housing_pivot.plot();
```

```python
ca_housing_markets = ca_housing_pivot.loc[:, ['San Francisco',
                                             'Los Angeles',
                                             'San Diego',
                                             'CA_average']]

ca_housing_markets.head()
```

```python
fig, ax = plt.subplots()

ax.plot(ca_housing_markets)

#ax.plot(ca_housing_markets['San Franciso'])
#ax.plot(ca_housing_markets['Los Angeles'])

ax.set_title('Housing Prices in CA Markets 2017-2022', fontsize = 16)
ax.set_xlabel('Year', fontsize = 12)
ax.set_ylabel("Price (Millions)", fontsize = 12);
```

```python
fig, ax = plt.subplots()

#ax.plot(ca_housing_markets)

ax.plot(ca_housing_markets.index, 
        ca_housing_markets['San Francisco'],
        label = 'San Francisco')


ax.plot(ca_housing_markets.index, 
        ca_housing_markets['Los Angeles'],
        label = 'Los Angeles') #have to add labels to get values for the legend


fig.suptitle('San Francisco Prices Decline in 2020', fontsize = 14)

ax.set_title('While LA Prices Keep Rising', fontsize = 11)
ax.set_xlabel('Year', fontsize = 12)
ax.set_ylabel("Median Home Price (Millions)", fontsize = 12
             )
ax.legend(['SF', 'LA'], loc = 'upper left', ncol = 2, frameon = False); 

#use bbox_to anchor = (#,#) to place outside the box 1 and above
#ncol will give horizontal list 
#frameon is for the box around the legend
```
