```python
data = [x * 10 for x in range(10)]
data
```

```python
labels = [f'{x}!' for x in range(10)]
labels
```

```python
plt.plot(labels, data); # ; at end line will supress the text like following:[<matplotlib.lines.Line2D at 0x21b7844c550>]
```

```python
plt.plot(labels, data)
plt.show() # use this to supress
```

```python
df = pd.DataFrame(data, labels).assign(series2 = lambda x: x[0] / 2)
df.head()
```

```python
plt.plot(df);
```

```python
#PyPlot API is all of the above
```

```python
#Object Oriented option
fig = plt.figure() #create empty canvas

ax = fig.add_subplot() #add figure to canvas
ax.plot(df);
```

```python
#better way

fig, ax = plt.subplots()
ax.plot(df)
fig.suptitle('Overall Title')
ax.set_title('Chart Title')
```

```python
plt.plot(df)

plt.title('Title');
```

```python
data_dict = {
    'Sales': [10,20,30,40],
    'Profits': [2,4,3,1],
    'Date': ['2022-01-01', '2022-02-01', '2022-03-01', '2022-04-01']
}

df = pd.DataFrame(
    data_dict,
    columns = ['Sales', 'Profits'],
    index = data_dict['Date']
)
df
```

```python
fig, ax = plt.subplots()

ax.plot(df)
ax.set_title('Product Sales and Profits');
```

```python
fig, ax = plt.subplots()

ax.plot(df.index, df['Sales']) #can split the df into series if want to format each line separately
ax.plot(df.index, df['Profits'], ls = '--')
ax.set_title('Product Sales and Profits');
```
